```python
def wordopt(text):
    text = text.lower()                               #converting into lower case
    text = re.sub('\[.*?\]', '', text)                #removing square braces and their contents
    text = re.sub("\\W"," ",text)                     #replacing non alphanumeric char with spaces
    text = re.sub('https?://\S+|www\.\S+', '', text)  #removing URLs
    text = re.sub('<.*?>+', '', text)                 #removing tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)     #removing punctuation
    text = re.sub('\n', '', text)                                       #removing newline char
    text = re.sub('\w*\d\w*', '', text)                                 #removing words containg digits
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
import nltk
nltk.download('stopwords')

```

```python

def wordopt2(text):
    text = text.lower()
    text = text.split()
    text = [ps.stem(word) for word in text if not word in stopwords.words('english')]
    text = ' '.join(text)
    return text
```

```python
data["text"] = data["text"].apply(wordopt2)
```

```python
data.head(10)
```

```python
#defining dependent and independent variables
x = data["text"]
y = data["class"]
```

```python
#splitting training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
#converting text to vector
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
#decision tree classification
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
#Random Forest Model
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train, y_train)
```

```python
pred_rfc = RFC.predict(xv_test)
```

```python
RFC.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rfc))
```

```python
#model testing
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_def_test["text"] = new_def_test["text"].apply(wordopt2)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction:{} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),output_lable(pred_DT[0]),output_lable(pred_RFC[0])))
```
